```python
names = {"MHKOOP":"Home owners","MAUT1":"1 car","MAUT2":"2 cars","MAUT0":"No car","MZFONDS":"National Health Service","MZPART":"Private health insurance","MINKM30":"Income < 30.000","MINK3045":"Income 30-45.000","MINK4575":"Income 45-75.000","MINK7512":"Income 75-122.000","MINK123M":"Income >123.000","MINKGEM":"Average income"}
# names
```

```python
long_names=[names[i] for i in names]
# long_names
```

```python
our_data.columns=long_names
our_data
```

```python
#kollo lists
def support(names,numbers):
    our_data2=our_data.copy()
    for i,j in zip(names,numbers):
        our_data2=our_data2.loc[our_data2[i]==j]
#     print(our_data2)
    return our_data2.count()[0]/5822
```

```python
#kollo lists
def convidence(names1,numbers1,names2,numbers2):
    return (support(np.concatenate((names1,names2)),np.concatenate((numbers1,numbers2)))/support(names1,numbers1))
```

```python
#kollo lists
def lift(names1,numbers1,names2,numbers2):
    return (support(np.concatenate((names1,names2)),np.concatenate((numbers1,numbers2)))/(support(names1,numbers1)*support(names2,numbers2)))

#kollo lists
def leverage(names1,numbers1,names2,numbers2):
    return (support(np.concatenate((names1,names2)),np.concatenate((numbers1,numbers2)))-(support(names1,numbers1)*support(names2,numbers2)))
```

```python
def printt(list1,list2):
    text=""
    for i,j in zip(list1,list2):
        text+=i+" ("+str(j)+") , "
    return text[:-2]
```

```python
index=[]
for i in long_names:
    unique=our_data[i].unique()
    index.append(sorted(unique))
print("index are : ")
index
```

```python
import itertools
all_combinations = []
for r in range(1,(len(long_names) + 1)):
    combinations_object = itertools.combinations(long_names, r)
    combinations_list = np.array(list(combinations_object))
#     print(combinations_list)
    all_combinations.append(combinations_list)

# print(len(all_combinations))
```

```python
s=1
for i in all_combinations[0:]:
    lists1=[]
    lists2=[]
    index2=[[],[],[],[],[],[],[],[],[],[],[],[]]
    for j in i:
        #j=['Home owners' '1 car' '2 car']
#         print(j)
        array=[]
        for k in j:
            l=long_names.index(k)
            array.append(index[l])
#         print(array)
        n = np.array(list(itertools.product(*array)))
        for m in n:
#             print(j)
#             print(m)
            supports=support(j,m)
            if supports>min_support:
                lists1.append(j)
                lists2.append(m)
                for p,o in zip(j,m):
                    l=long_names.index(p)
                    if o not in index2[l]:
                        index2[l].append(o)
    print(f"index after ( {s}-itemsets ) with support >= min_support: ")
    print(index2)
    print("")
    s+=1
#     print(len(lists1))
    if len(lists1)!=0:
        listsx=lists1
        listsy=lists2
        index=index2
    else:
        break
```
